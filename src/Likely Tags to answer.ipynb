```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import wordcloud
```

```python
questionTags= pd.read_csv('data/tagWiseAnswerCountPerYear.csv',low_memory=False)
```

```python
#questionTags.plot(x='tag_name',y='AnswerCount',kind ='bar')

questionTags= questionTags.dropna()
```

```python
plt.scatter(questionTags.tag_name,questionTags.AnswerCount,c=questionTags.YEAR)
```

```python
tagwise = questionTags[['tag_name','AnswerCount']].groupby('tag_name').sum()
```

```python
tagwise.head()
#id            owner_user_id
#questionId    userId
```

```python
top10tags= tagwise.sort_values('AnswerCount',ascending=False).head(10)
```

```python
top10tags.plot(kind ='bar',title ='Top 10 Tags Likely to answer',)
```
