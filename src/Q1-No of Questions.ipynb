```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
import seaborn as sns;
import matplotlib.pyplot as plt
import wordcloud;
import numpy as np ;
```

```python
allQuestions=pd.read_csv('allquestions.csv', low_memory=False);
allQuestions
```

```python
allUsers=pd.read_csv('allusers.csv', low_memory=False);
allUsers
```

```python
pakUsers=allUsers[allUsers['location'].str.contains('Pakistan')]
pakUsers
```

```python
pakQuestions=pd.merge(allQuestions,pakUsers,left_on='owner_user_id',right_on='id',how='inner')
pakQuestions
```

```python
indiaUsers=allUsers[allUsers['location'].str.contains('India')]
indiaUsers
```

```python
indiaQuestions=pd.merge(allQuestions,indiaUsers,left_on='owner_user_id',right_on='id',how='inner')
indiaQuestions
```

```python
usaUsers=allUsers[allUsers['location'].str.contains('USA')|allUsers['location'].str.contains('United States')]
usaUsers
```

```python
usaQuestions=pd.merge(allQuestions,usaUsers,left_on='owner_user_id',right_on='id',how='inner')
usaQuestions
```

```python
pakGroupedQuesitons = pakQuestions.groupby('year')
groupedQuesitonsCount=pakGroupedQuesitons['year'].agg(['count'])
groupedQuesitonsCount.rename(columns={'count':'Pakistan'}, 
                 inplace=True)
groupedQuesitonsCount
```

```python
indiaGroupedQuesitons = indiaQuestions.groupby('year')
groupedQuesitonsCount['India']=indiaGroupedQuesitons['year'].agg(['count'])
groupedQuesitonsCount
```

```python
usaGroupedQuesitons = usaQuestions.groupby('year')
groupedQuesitonsCount['USA']=usaGroupedQuesitons['year'].agg(['count'])
groupedQuesitonsCount
```

```python
sns.set()
ax=sns.lineplot( markers=True, dashes=False,data=groupedQuesitonsCount)
ax.set(xlabel='Years', ylabel='No of Questions')
```
