```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
import seaborn as sns;
import matplotlib.pyplot as plt
import wordcloud;
import numpy as np ;
```

```python
def generateWordCloud(df,column,figName):
    allTags = []
    for _, r in df.iterrows():

        splited = str(r[column]).split('|')
        allTags += splited

    cloudTags = ' '.join(allTags).lower()
    cloud = wordcloud.WordCloud(background_color='white',
                            max_font_size=200,
                            width=1600,
                            height=800,
                            relative_scaling=.5,
                            collocations=False).generate(cloudTags)
    plt.figure(figsize=(20,10))
    plt.axis('off')
    plt.savefig(figName)
    plt.imshow(cloud);
```

```python
allQuestions=pd.read_csv('allquestions.csv', low_memory=False);
```

```python
generateWordCloud(allQuestions,'tags','allquestions.png')
```
